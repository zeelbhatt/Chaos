```python
import matplotlib.pyplot as plt
import math
import numpy as np
```

```python
x = 0.5
r = 1.3

time = [0]
population = [0.9]


for i in range(0,15):
    x = r*x*(1-x)
    #x = r*math.sin(x)
    time.append(i)
    population.append(x)
    
print("Population is ", x)

eq = (r-1)/r

print("Equilibrium is ", eq)


plt.plot(time, population, 'bo', time, population, 'k')

plt.ylim(0,1)
plt.show()
```

```python
x = 0.9
r = 1

rate = []
eqb = []

#time = [0]
#population = [0.9]

for j in range(0, 1000000):
    
    for k in range(0, 50):
        x = x*r*(1-x)
    
    eqb.append(x)
    rate.append(r)
    
    r = r + 0.001
    
plt.plot(rate, eqb, 'ko', markersize = 3)
plt.ylim(0,1)
plt.xlim(0,6)
```

```python
def f(t):
    return (t-1)/t

t = np.arange(1, 6.0, 0.01)
plt.plot(t, f(t), 'ko', markersize = 3)
plt.xlim(0,6)
plt.ylim(0,1)
```
